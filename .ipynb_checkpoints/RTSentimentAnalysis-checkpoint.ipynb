```python
#Imports
import pandas as pd
import nltk
import os
```

```python
posTrain = './aclImdb/train/pos'
negTrain = './aclImdb/train/neg'
url_posTrain = './aclImdb/train/urls_pos.txt'
url_negTrain = './aclImdb/train/urls_neg.txt'
posTrainUrls = ''
negTrainUrls = ''

with open(url_posTrain, 'r', encoding='utf-8') as myfile:
        posTrainUrls = myfile.read().split('\n')
        
with open(url_negTrain, 'r', encoding='utf-8') as myfile:
        negTrainUrls = myfile.read().split('\n')

posTrainFiles = os.listdir(posTrain)
posTrainArray = []
for file in posTrainFiles:
    dict_ = {}
    url = posTrain + '/{}'.format(file)
    dict_['id'] = int(file.split('_')[0])
    dict_['url'] = posTrainUrls[int(file.split('_')[0])]
    dict_['grade'] = file[:-4].split('_')[1]
    with open(url, 'r', encoding='utf-8') as myfile:
        dict_['review'] = myfile.read()
    posTrainArray.append(dict_)


negTrainFiles = os.listdir(negTrain)
negTrainArray = []
for file in negTrainFiles:
    dict_ = {}
    url = negTrain + '/{}'.format(file)
    dict_['id'] = int(file.split('_')[0])
    dict_['url'] = negTrainUrls[int(file.split('_')[0])]
    dict_['grade'] = file[:-4].split('_')[1]
    with open(url, 'r', encoding='utf-8') as myfile:
        dict_['review'] = myfile.read()
    negTrainArray.append(dict_)
        
```

```python
posTrainDF = pd.DataFrame(posTrainArray).sort_values('id')
display(posTrainDF.set_index('id'))
```
